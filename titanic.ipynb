# Data Cleaning Task

In this task, we will be cleaning and preparing the data for analysis. Data cleaning is an important step in the data analysis process as it helps to ensure the accuracy and reliability of the data.

## Steps for Data Cleaning

1. **Import the necessary libraries**: Before we start cleaning the data, we need to import the required libraries such as pandas, numpy, and matplotlib.

2. **Load the dataset**: We will load the dataset into a pandas dataframe using the `read_csv()` function.

3. **Explore the data**: It is important to understand the structure and content of the dataset. We will examine the columns, data types, and any missing values in the dataset.

4. **Handle missing values**: Missing values can affect the analysis results. We will identify and handle missing values by either removing the rows or imputing the missing values.

5. **Remove duplicates**: Duplicates can skew the analysis results. We will identify and remove any duplicate rows in the dataset.

6. **Correct data types**: Sometimes, the data types of certain columns may not be appropriate. We will convert the data types of columns if necessary.

7. **Clean and transform data**: We will clean and transform the data as per the requirements of the analysis. This may involve removing outliers, standardizing values, or creating new variables.

8. **Validate the cleaned data**: After cleaning the data, it is important to validate the results. We will perform checks and tests to ensure the data is clean and ready for analysis.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Set the maximum number of columns to be displayed to 10
pd.set_option('display.max_columns', 10)
# Set the maximum width to a large value
pd.set_option('display.width', 1000)
```

```python
titanic = pd.read_csv('Titanic-Dataset.csv')
titanic.head()
```

```python
"""
Cell generated by Data Wrangler.
"""

def clean_data(titanic):
    # Change column type to bool for column: 'Survived'
    titanic = titanic.astype({'Survived': 'bool'})
    # Drop column: 'PassengerId'
    titanic = titanic.drop(columns=['PassengerId'])
    # Drop column: 'Ticket'
    titanic = titanic.drop(columns=['Ticket'])
    # Split text using string ',' in column: 'Name'
    loc_0 = titanic.columns.get_loc('Name')
    titanic_split = titanic['Name'].str.split(pat=',', expand=True).add_prefix('Name_')
    titanic = pd.concat([titanic.iloc[:, :loc_0], titanic_split, titanic.iloc[:, loc_0:]], axis=1)
    titanic = titanic.drop(columns=['Name'])
    # Remove leading and trailing whitespace in column: 'Name_0'
    titanic['Name_0'] = titanic['Name_0'].str.strip()
    # Split text using string '.' in column: 'Name_1'
    loc_0 = titanic.columns.get_loc('Name_1')
    titanic_split = titanic['Name_1'].str.split(pat='.', expand=True).add_prefix('Name_1_')
    titanic = pd.concat([titanic.iloc[:, :loc_0], titanic_split, titanic.iloc[:, loc_0:]], axis=1)
    titanic = titanic.drop(columns=['Name_1'])
    # Drop column: 'Name_1_0'
    titanic = titanic.drop(columns=['Name_1_0'])
    # Drop column: 'Name_1_2'
    titanic = titanic.drop(columns=['Name_1_2'])
    # Drop column: 'Cabin'
    titanic = titanic.drop(columns=['Cabin'])
    # Remove leading and trailing whitespace in column: 'Name_1_1'
    titanic['Name_1_1'] = titanic['Name_1_1'].str.strip()
    # Rename column 'Name_0' to 'Last Name'
    titanic = titanic.rename(columns={'Name_0': 'Last Name'})
    # Rename column 'Name_1_1' to 'First Name'
    titanic = titanic.rename(columns={'Name_1_1': 'First Name'})
    return titanic

titanic_clean = clean_data(titanic.copy())
titanic_clean.head()
```

## Data Cleaning Completed

The data has been successfully cleaned and prepared for analysis. All necessary steps for data cleaning, including handling missing values, removing duplicates, correcting data types, and cleaning and transforming the data, have been performed.

The cleaned dataset is ready for further analysis and can be used with confidence.

```python
# Summary Statistics
summary_stats = titanic_clean.describe()
print(summary_stats)

# Data Visualization
plt.figure(figsize=(6, 4), dpi=200)
plt.style.use('ggplot')
titanic_clean['Age'].hist()
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
plt.clf()

# Correlation Analysis
plt.figure(figsize=(6, 4), dpi=200)
titanic_corr = titanic_clean[['Age', 'Fare', 'Survived', 'SibSp', 'Parch', 'Pclass']].corr()
sns.heatmap(titanic_corr, annot=False)
plt.show()
plt.clf()

# Outlier Detection
plt.figure(figsize=(6, 4), dpi=200)
sns.boxplot(x=titanic_clean['Fare'])
plt.show()
plt.clf()

# Data Distribution
plt.figure(figsize=(6, 4), dpi=200)
sns.histplot(titanic_clean['Age'], kde=True, stat="density")
plt.show()
plt.clf()

# Data Quality Check
missing_values = titanic_clean.isnull().sum()
print(missing_values)

```

```python
sns.pairplot(titanic_clean, hue='Survived')
plt.show()
```

```python
# Calculate the count and percentage of women and men who survived
survived_count = titanic_clean[titanic_clean['Survived'] == 1].groupby('Sex')['Survived'].count()
total_count = titanic_clean.groupby('Sex')['Survived'].count()
percentage = (survived_count / total_count) * 100

# Plot the bar chart
plt.figure(figsize=(6, 4), dpi=200)
plt.bar(percentage.index, percentage.values)
plt.xlabel('Sex')
plt.ylabel('Percentage of Survived')
plt.title('Percentage of Survived by Sex')
plt.show()
```

```python
women_count = titanic_clean['Sex'].value_counts()['female']
men_count = titanic_clean['Sex'].value_counts()['male']
porcent_men = survived_count['male']/ men_count*100
porcent_women = survived_count['female']/women_count*100
plt.figure(figsize=(6, 4), dpi=200)
plt.bar(['Male','Female'],[porcent_men,porcent_women])
plt.xlabel('Sex')
plt.ylabel('Percentage of Survived')
plt.show()
print(porcent_men)
print(porcent_women)
```

```python
male_mean_age = titanic_clean[titanic_clean['Sex'] == 'male']['Age'].mean()
female_mean_age = titanic_clean[titanic_clean['Sex'] == 'female']['Age'].mean()

titanic_clean.loc[titanic_clean['Sex'] == 'male', 'Age'] = titanic_clean.loc[titanic_clean['Sex'] == 'male', 'Age'].fillna(male_mean_age)
titanic_clean.loc[titanic_clean['Sex'] == 'female', 'Age'] = titanic_clean.loc[titanic_clean['Sex'] == 'female', 'Age'].fillna(female_mean_age)

titanic_clean['Age'].isnull().sum()
titanic_clean

```

```python
from pywaffle import Waffle
import matplotlib.patches as mpatches

# Count the values of the "Embarked" column
embarked_counts = titanic_clean['Embarked'].value_counts()

# Specify the colors for the legend
colors = ['#FFA07A', '#87CEFA', '#98FB98']

# Create a waffle chart with specified colors

fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    columns=10,
    values=embarked_counts,
    colors=colors,
    figsize=(6, 4), 
    dpi=200,
)
patches = [mpatches.Patch(color=color, label=label) for color, label in zip(colors, embarked_counts.keys())]
# Set the title
plt.legend(handles=patches, bbox_to_anchor=(1, 1))
plt.title('Distribution of Embarked')

# Show the chart
plt.show()
```
